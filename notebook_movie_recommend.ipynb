```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('TMDBDataSet/tmdb_5000_movies.csv')
credits = pd.read_csv('TMDBDataSet/tmdb_5000_credits.csv') 
```

```python
movies.sample(6)
```

```python
credits.sample(5)
```

```python
# merging dfs
movie = movies.merge(credits, left_on=['id', 'title'], right_on=['movie_id', 'title'], how='inner', validate="1:1")
```

```python
movie.sample()
```

```python
req_columns = ["genre", "movie_id", "keywords", "title", "overview", "cast", "crew"]
```

```python
movies = movie[["movie_id", "genres", "keywords", "title", "overview", "cast", "crew"]]
```

```python
# Remove missing data
```

```python
movies.isnull().sum()
```

```python
movies = movies.dropna()
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    list_of_values = []
    for i in ast.literal_eval(obj):
        list_of_values.append(i['name'])
    return list_of_values
        
```

```python
movies.sample(5)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.sample(5)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
import ast
def convert_cast(obj):
    list_of_values = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            list_of_values.append(i['name'])
            counter+=1
        else:
            break;
    return list_of_values
        
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.sample(5)
```

```python
import ast
def fetch_director(obj):
    list_of_values = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list_of_values.append(i['name'])
            break;
    return list_of_values
        
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.sample(5)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.sample(5)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
# movies.iloc[0]['tags']
```

```python
movies.columns
```

```python
df = movies[['movie_id', 'title', 'tags' ]]
```

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
```

```python
df.sample(5)
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df.sample()
```

```python
# apply stemming to get the root of the word
# !pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df["tags"] =  df["tags"].apply(stem)
```

```python
df['tags'][0]
```

```python
# Perform Vectorization
# concatinate all the tags---> apply bagofwords-->exclude stop words
#
```

```python
# !pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
print(cv.get_feature_names_out())
```

```python

# calculate cosine similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# sorted(similarity[0], reverse=True)
# problem--> we have to retain the index
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
def  recommend(movie):
    movie_index = df[df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(df.iloc[i[0]]['title'])
        # print(i[0])
    # return 
```
